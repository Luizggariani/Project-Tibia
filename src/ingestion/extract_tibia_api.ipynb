```python
import requests
import json

url = 'https://api.tibiadata.com/v4/creatures'

response = requests.get(url)

with open('creatures.json', 'w') as file:
    json.dump(response.json(), file)
```

```python
response_json = response.json()

print(response_json['creatures'])
```

```python
creature_list = response_json['creatures']['creature_list']
```

```python
i = 0
for single_creature in creature_list:
    i = i + 1
    if 'Dragon' in single_creature['name']:
        print(f"o {single_creature['name']} esta em posição: {i}")
```

```python
import pandas as pd

df = pd.DataFrame(creature_list)

df.head()
```
